#Data Cleaning
*   Remove empty rows
*   Remove duplicate text
*   Remove very short sentences
*   Remove special symbols / HTML
*   Normalize Hindi text (extra spaces, line breaks)

```python
from datasets import load_dataset
import pandas as pd

# load dataset
dataset = load_dataset("marsh-mellow/hindi_wikipedia")

# convert to pandas
df = pd.DataFrame(dataset["train"])

# save as CSV
df.to_csv("hindi_wikipedia.csv", index=False)
```

```python
import pandas as pd
import re

# load CSV created in step 1
df = pd.read_csv("/content/hindi_wikipedia.csv")

print("Before cleaning:", df.shape)
print(df.head())
```

```python
df.head()
```

```python
print(df['text'][0])
```

```python
# check which rows have empty or null text
empty_rows = df[df["text"].isna() | (df["text"].str.strip() == "")]
print(empty_rows)
```

```python
# to see only row numbers
print(empty_rows.index.tolist())
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["text"])
```

```python
# find duplicate rows in 'text' column
duplicate_rows = df[df.duplicated(subset=["text"])]
```

```python
print(duplicate_rows)
print("Number of duplicate rows:", duplicate_rows.shape[0])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(subset=["text"])
```

```python
# check rows with text length < 20 characters
short_rows = df[df["text"].str.len() < 20]
print("Number of short rows:", short_rows.shape[0])
print(short_rows.head())
```

```python
# remove rows with text length < 20
df = df[df["text"].str.len() >= 20]
```

```python
import re

# check rows that contain HTML tags or special (non-Hindi) symbols
pattern = r"<.*?>|[^\u0900-\u097F\s]"

rows_with_symbols = df[df["text"].str.contains(pattern, regex=True, na=False)]

print("Rows with HTML/special symbols:", rows_with_symbols.shape[0])
print(rows_with_symbols.head())
```

```python
def clean_text(text):
    if pd.isna(text):
        return ""

    text = re.sub(r"<.*?>", "", text)          # remove HTML tags
    text = re.sub(r"http\S+", "", text)        # remove links
    text = re.sub(r"[^\u0900-\u097F\s]", " ", text)  # keep Hindi chars only
    text = re.sub(r"\s+", " ", text)           # remove extra spaces
    text = text.strip()

    return text

df["text"] = df["text"].apply(clean_text)
```

```python
df.head()
```

```python
# check rows with extra spaces or line breaks
rows_with_spaces = df[df["text"].str.contains(r"\s{2,}|\n|\t", regex=True, na=False)]
print("Rows with extra spaces/line breaks:", rows_with_spaces.shape[0])
print(rows_with_spaces.head())
```

```python
# remove extra spaces and line breaks
df["text"] = df["text"].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
df = df.reset_index(drop=True)

df.to_csv("hindi_cleaned_final.csv", index=False)

print("Cleaned dataset saved as hindi_cleaned_final.csv")
print("Final shape:", df.shape)
print(df.head())
```
